```python
import pandas as pd
```

```python
df = pd.read_csv("BA_reviews.csv")
df.head()
```

```python
df.info()
```

```python
df = df.drop("Unnamed: 0", axis=1)
df.head()
```

```python
df['verified'] = df["reviews"].str.contains("Trip Verified")
df.head()
```

```python
df.verified = df.verified.replace(True,"Yes")
df.verified = df.verified.replace(False,"No")
df.head()
```

```python
dfvy = df["reviews"].str.strip("✅ Trip Verified |")
dfvn = dfvy.str.strip("Not Verified |")
df["reviews"] = dfvn
df.head()
```

```python
df.date = pd.to_datetime(df.date)
df.head()
```

```python
df.country.value_counts()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

df['Sentiment'] = df['reviews'].apply(lambda x : analyzer.polarity_scores(x)['compound'])


def PNN(y):
    if y >= 0.5:
        return 'Positive'
    elif y < 0 :
        return 'Negative'
    else:
        return 'Neutral'
    
df['Analysis'] = df['Sentiment'].apply(PNN)
df.head()
```

```python
df.Analysis.value_counts()
```

```python
df.to_csv("BA_reviews.csv")
```
